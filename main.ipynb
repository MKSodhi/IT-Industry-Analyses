```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('Data/employee_data.csv')

df.head()
```

## Verificando qualidade dos dados:

```python
print(df.dtypes)
```

```python
# Count and print the number of null values in each column
# Conta e imprime o número de valores nulos em cada coluna
null_counts = df.isna().sum()

print('Contagem de valores nulos por coluna:')
print(null_counts[null_counts > 0])
```

## Distribuição Salarial por Gênero:

```python
# Agrupando os dados por gênero e calculando a média salarial
salary_distribution = df.groupby('Gender')['Salary'].mean().reset_index()

print(salary_distribution)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Gender', y='Salary', data=df, estimator=sum, palette=['pink', 'blue'])
plt.xlabel('Gender')
plt.ylabel('Total Salary')
plt.title('Salary Distribution by Gender')
plt.show()
```

## Relação entre Experiência e Salário:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Experience (Years)', y='Salary', hue='Gender', data=df, palette=['pink', 'blue'], s=100)
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.title('Relationship between Experience and Salary')

plt.show()
```

## Média Salarial por Posição:

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Salary', y='Position', data=df, estimator='mean', palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Position')
plt.title('Average Salary by Position')
plt.xticks(rotation=45)

plt.show()
```

## Diferença Salarial entre Gêneros em Cada Posição:

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Salary', y='Position', hue='Gender', data=df, palette=['pink', 'blue'])
plt.xlabel('Position')
plt.ylabel('Salary')
plt.title('Salary Difference by Gender in Each Position')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

## Salário Médio por Anos de Experiência:

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Experience (Years)', y='Salary', data=df, marker='o', color='b')
plt.xlabel('Years of Experience')
plt.ylabel('Average Salary')
plt.title('Average Salary by Years of Experience')
plt.grid(True)

plt.show()
```

## Distribuição de Funcionários por Gênero e Posição:

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Position', hue='Gender', data=df, palette=['pink', 'blue'])
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.title('Distribution of Employees by Gender and Position')
plt.xticks(rotation=45)
plt.legend(title='Gender')

plt.show()
```

## Salário por Posição para Diferentes Níveis de Experiência:

```python
df_0_10 = df[df['Experience (Years)'] <= 10]
df_11_20 = df[(df['Experience (Years)'] > 10) & (df['Experience (Years)'] <= 20)]
```

```python
plt.figure(figsize=(12, 6))
sns.pointplot(x='Position', y='Salary', hue='Experience (Years)', data=df_0_10, palette='coolwarm')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.title('Salary by Position for 0 to 10 Years of Experience')
plt.xticks(rotation=45)
plt.legend(title='Experience (Years)')

plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.pointplot(x='Position', y='Salary', hue='Experience (Years)', data=df_11_20, palette='coolwarm')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.title('Salary by Position for 11 to 20 Years of Experience')
plt.xticks(rotation=45)
plt.legend(title='Experience (Years)')

plt.show()
```

## Identificação de Outliers Salariais:

```python
# Calculando Q1 (1º quartil) e Q3 (3º quartil)
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)

# Calculando o intervalo interquartil (IQR)
IQR = Q3 - Q1

# Definir os limites para outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar os outliers
outliers = df[(df['Salary'] < lower_bound) | (df['Salary'] > upper_bound)]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Salary', data=df, color='lightblue')
plt.xlabel('Salary')
plt.title('Boxplot of Salaries with Outliers Highlighted')
plt.scatter(outliers['Salary'], [1] * len(outliers), color='red', label='Outliers', zorder=5)
plt.legend()

plt.show()
```

## Impacto do Gênero e da Experiência Combinados no Salário:

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Experience (Years)', y='Salary', hue='Gender', size='Experience (Years)', sizes=(50, 300), palette='coolwarm', data=df, legend='full')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.title('Impact of Gender and Experience on Salary')
plt.grid(True)
plt.legend(title='Gender', loc='upper left')

plt.show()
```
