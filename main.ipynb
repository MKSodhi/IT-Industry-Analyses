```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('Data/employee_data.csv')

df.head()
```

## Verificando qualidade dos dados:

```python
print(df.dtypes)
```

```python
# Count and print the number of null values in each column
# Conta e imprime o número de valores nulos em cada coluna
null_counts = df.isna().sum()

print('Contagem de valores nulos por coluna:')
print(null_counts[null_counts > 0])
```

## Distribuição Salarial por Gênero:

```python
# Agrupando os dados por gênero e calculando a média salarial
salary_distribution = df.groupby('Gender')['Salary'].mean().reset_index()

print(salary_distribution)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Gender', y='Salary', data=df, estimator=sum, palette=['pink', 'blue'])
plt.xlabel('Gender')
plt.ylabel('Total Salary')
plt.title('Salary Distribution by Gender')
plt.show()
```

## Relação entre Experiência e Salário:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Experience (Years)', y='Salary', hue='Gender', data=df, palette=['pink', 'blue'], s=100)
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.title('Relationship between Experience and Salary')

plt.show()
```

## Média Salarial por Posição:

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Salary', y='Position', data=df, estimator='mean', palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Position')
plt.title('Average Salary by Position')
plt.xticks(rotation=45)

plt.show()
```

## Diferença Salarial entre Gêneros em Cada Posição:

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Salary', y='Position', hue='Gender', data=df, palette=['pink', 'blue'])
plt.xlabel('Position')
plt.ylabel('Salary')
plt.title('Salary Difference by Gender in Each Position')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```
